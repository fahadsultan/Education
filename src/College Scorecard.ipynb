# College Scoreboard Dataset

This notebook file takes a preliminary look at the **[College Scoreboard Dataset](https://collegescorecard.ed.gov/)** collected by the US Department of Education. 

The dataset, indexed by institutions, as well as field of study, includes all the information that is available on the [website]((https://collegescorecard.ed.gov/).

The dataset includes **2989 indicators** for each institution. 

The data is also longitudinal: available for multiple years. 

_Warning: a lot of them, particularly ones broken down by field of study, are unavailable for most institutes._

Some indicators of particular interest include: 


* Costs 

    * **Average Annual Cost**
    * **Average Annual Cost, by Family Income**


* Graduation and Retention

    * **Graduation Rates**
    * **Graduation Rates, Pell Grant Recepients only** 

    * **Outcomes, 8 years after attending**: graduated, transfered, withdrew
    * **Students who return after first year** (percentage)


* Financial Aid and Debt 

    * **Students receiving Federal Loans** (percentage)
    * **Median Total Debt after graduation**
    * **Typical Monthly Loan Payment**
    * **Repayment rate**


* Typical Earnings 

    * **Median Earnings** : The median annual earnings of individuals that received federal student aid and began college at this institution 10 years ago, regardless of their completion status.

    * **Percentage Earning More than a High School Graduate**
    
    * **Earnings after Completing Field of Study**


* Campus Diversity 

    * **Socio Economic Diversity**
    * **Faculty to Student Ratio**
    * **Race/Ethnicity** of Student Body and Full Time Staff


* Test Scores and Acceptance 
    * **SAT Scores**: Math and Critical Reading 
    * **ACT**

## Correlations between Ranking and Scoreboard indicators

```python
ranking = pd.read_excel('../data/US Higher Education_ Value vs. Vulnerability.xlsx', skiprows=5)
ranking = ranking[ranking['Ranking List']=='USN National College Ranking']
ranking = ranking[['Institution Name', 'Rank']]
```

```python
merged = pd.merge(data, ranking, left_on='INSTNM', right_on='Institution Name')
```

```python
cols = list(merged.dtypes[merged.dtypes!='object'].index)
furman = merged[merged['Institution Name']=='Furman University']
for col in cols:
    if col in ['Rank', 'Institution Name', 'name']:
        continue
    plt.scatter(merged['Rank'], merged[col])
    
    topidx = merged[col].argmax()
    if topidx>=0:
        top = merged.loc[topidx]
        plt.title('Top %s (rank: %s)' % (top['Institution Name'], top['Rank']))
    plt.scatter(furman['Rank'], furman[col])
    plt.xlabel('Rank')
    plt.ylabel(col)
    plt.legend(['Other', 'Furman'], loc='upper right')
    plt.savefig('../out/scoreboard/%s' % col)
    plt.close()
```
